```python
import pandas as pd
import numpy as np

# Create 100 rows of fake student data
np.random.seed(42)
n_rows = 100

data = {
    'student_id': range(1, n_rows + 1),
    'chapter_order': np.random.randint(1, 6, n_rows),
    'time_spent': np.random.randint(5, 120, n_rows),
    'score': np.random.randint(20, 100, n_rows),
}

# Logic: If score > 60 and time > 30, they likely completed (1), else (0)
df = pd.DataFrame(data)
df['completion_status'] = ((df['score'] > 60) & (df['time_spent'] > 30)).astype(int)

df.to_csv('student_data.csv', index=False)
print("New student_data.csv with 100 rows created!")
```
