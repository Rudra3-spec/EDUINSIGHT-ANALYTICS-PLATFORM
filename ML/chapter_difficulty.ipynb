```python
import pandas as pd
import json

# 1. Load & Clean
df = pd.read_csv('student_data.csv')

# 2. Statistical Analysis (The "Logic")
# We group by chapter to see averages
difficulty_report = df.groupby('chapter_order').agg({
    'completion_status': 'mean', # Low mean = High difficulty
    'score': 'mean',             # Low score = High difficulty
    'time_spent': 'mean'         # High time = High difficulty
}).reset_index()

# 3. Calculation
# A simple formula to rank difficulty 0-100
difficulty_report['difficulty_score'] = (1 - difficulty_report['completion_status']) * 100

# 4. Save for MERN Dashboard
difficulty_report.to_csv('chapter_difficulty.csv', index=False)
print(difficulty_report)

```
